## K-means clustering

```python
cdata=clustdata[['CustomerID', 'combgroup5']]
cdata
```

```python
data1=data[['InvoiceNo', 'CustomerID', 'StockCode']]
data1
```

```python
cdata
```

```python
mergeDataAndClustdata = (data1.merge(cdata, left_on='CustomerID', right_on='CustomerID')
          .reindex(columns=['InvoiceNo', 'CustomerID', 'StockCode', 'combgroup5']))
```

```python
mergeDataAndClustdata
```

```python
data2=mergeDataAndClustdata[['InvoiceNo', 'StockCode', 'combgroup5']]
data2
```

```python
data2.describe
```

```python
data2.dtypes
```

```python
type('ABS')
```

```python
data2 = data2.astype({'StockCode': 'str'})
```

```python
data2[['StockCode']].dtypes
```

```python
data2=data2.sort_values(by=['StockCode', 'combgroup5'], ascending=True)
```

```python
data2.head(20)
```

```python
n=5
for i in range(0, n):
    print(i)
    print(clustdata.loc[clustdata['combgroup'+str(n)] == i]['RFM'].unique())
```

```python
# vip-group 3 will come first
# # non-vip order-[1, 2, 4, 0]
```

```python
data2['combgroup5']=pd.Categorical(data2['combgroup5'], [3, 1, 2, 4, 0])
data2
```

```python
findata1=data2.sort_values(by=['StockCode', 'combgroup5'])
```

```python
findata1[findata1['StockCode']=='10002']
```

```python
# findata1['InvoiceNo'].split(findata['combgroup5'])
```

```python
findata1.head()
```

```python
# findata1.to_excel('grouped-data.xlsx')
```

```python
xtemp=pd.DataFrame({'count' : findata1.groupby( [ "StockCode", "combgroup5"] ).size()}).reset_index()
```

```python
# findata2=findata1.groupby(['StockCode', 'combgroup5']).size()
# findata2
```

```python
# x1=pd.DataFrame(findata2)
# x1
```

```python
xtemp
```

```python
x2=xtemp.pivot(index='StockCode', columns='combgroup5', values='count')
x2
```

```python
x2.to_csv('stock-group.csv', header=None)
```

```python
x2.to_csv('stock-group-edit.csv', header=None, index=False)
```
