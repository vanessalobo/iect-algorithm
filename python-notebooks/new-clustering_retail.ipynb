DBSCAN with R, F, M

```python
db_default = DBSCAN(eps = 0.002, min_samples = 100).fit(data[['R', 'F', 'M']]) 
labels = db_default.labels_ 
#print(labels)
data['dbscan']=labels
print(data['dbscan'].unique())
plt.scatter(data['RFM'], data['dbscan'], s=10, cmap='viridis')
plt.show()
```

## K-means hierachical - R, F, M individual

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
n=[3, 5, 8, 10]
for i in n:
    agglo=AgglomerativeClustering(n_clusters=i).fit(data[['R', 'F', 'M']])
    data['agg'+str(i)]=agglo.labels_
    print(data['agg'+str(i)].unique())
    plt.scatter(data['RFM'], data['agg'+str(i)], s=50, cmap='viridis')
    plt.show()
```
