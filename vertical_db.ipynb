**RFM CALC**

```
df=pd.read_excel('data2.xlsx')
df.head()
```

```
temp=df.groupby('CustomerID')['InvoiceNo'].nunique()
```

```
temp=temp.to_frame(name="Frequency").reset_index()
```

```
temp
```

```
d1=df.groupby('CustomerID')['InvoiceDate'].first().to_frame().reset_index()
d1
```

```
d1['InvoiceDate']
```

```
temp['Recency']=pd.Timestamp('2021-01-01') - d1['InvoiceDate']
```

```
temp
```

```
df['prod']=df['Quantity']*df['UnitPrice']
df
```

```
d2=df.groupby('CustomerID')['prod'].sum().to_frame().reset_index()
```

```
d2
```

```
temp['Monetary']=d2['prod']
temp
```

```
temp.head(10)
```

```
temp['Recency'].value_counts().head(10)
```

```
temp.to_excel('rfm.xlsx')
```
